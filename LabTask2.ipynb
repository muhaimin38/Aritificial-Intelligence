**Question 7**

```python
class Student:
    def __init__(self, sid, name, student_class):
        self.student_id = sid
        self.student_name = name
        self.student_class = student_class

    def display(self):
        for key, value in self.__dict__.items():
            print(key, ":", value)

s1 = Student(101, "Ali", "BSCS")
s1.display()
```

**Question 10**

```python
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height


t = Trapezoid(5, 6, 4)
p = Parallelogram(10, 3)

if t.area() > p.area():
    print("Trapezoid has greater area")
else:
    print("Parallelogram has greater area")
```

**Question 4**

```python
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


data = [
    (5, 6, 4),
    (7, 8, 5),
    (4, 5, 6)
]

for i in range(len(data)):
    a, b, h = data[i]
    t = Trapezoid(a, b, h)
    print("Area", i + 1, ":", t.area())
```
